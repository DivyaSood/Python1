```python
#importing of Python Modules Required
import csv
import matplotlib.pyplot as plt
'''%matplotlib inline''' #used to print graph in notebook    
import numpy as np
from sklearn import linear_model
import re
#Main Code
lst1=[]
lst2=[]
lst3=[]
lst4=[]
input_file=csv.DictReader(open("HousepricesinCalifornia2.csv"))
for line in input_file:
    lst1.append(line['Alameda'])
    lst2.append(line['Mon-Yr'])
    lst3.append(line['Butte'])
    lst4.append(line['Mon-Yr'])
print "House prices in Alameda=",lst1
print "House price in Butte=",lst3
lst5 = [re.sub("[a-zA-Z$,]", "", elem) for elem in lst1]
lst6 = [re.sub("[a-zA-Z$,]", "", elem) for elem in lst3]
print "House price in Alameda(without string)=",lst5
print "House price in Butte(without string)=",lst6
y1=np.array(lst5).astype(np.float)
y2=np.array(lst6).astype(np.float)
print y1,y2
lst7=[]
for x in range(len(lst2)):
    lst7.append(x)
print "month-yr of Alameda",lst7
lst8=[lst7]
print lst8
x1= np.asarray(lst8)
print x1
Matrix1=np.matrix.transpose(x1)
print "value of Matrix1= ",Matrix1
#Plotting of Graph
lm=linear_model.LinearRegression()
plt.subplot(2,1,1)
lm.fit(Matrix1,y1)
x_test=[[0],[200],[341],[365]]
y_predict=lm.predict(x_test)
plt.scatter(lst8,y1, color='blue')
plt.plot(x_test,y_predict,color='black',linewidth=2)
plt.title('House prices in Alameda ')
plt.xlabel('Mon-Yr(1990-2016)')
plt.ylabel('House prices($)')
plt.subplot(2,1,2)
lm.fit(Matrix1,y2)
x_lst=[[0],[200],[341],[365]]
y_predict=lm.predict(x_test)
plt.scatter(lst8,y2,color='red')
plt.plot(x_test,y_predict,color='black',linewidth=2)
plt.title('House prices in Butte ')
plt.xlabel('Mon-Yr(1990-2016)')
plt.ylabel('House prices ($)')
plt.show()
```

```python
def delete_column(self):

# Instantiating a Workbook object by excel file path
    workbook = self.Workbook(self.dataDir + 'HousepricesinCalifornia2')

# Accessing the first worksheet in the Excel file
    worksheet = workbook.getWorksheets().get(0)

# Deleting a column from the worksheet at 2nd position
    worksheet.getCells().deleteColumns(1,1,True)

# Saving the modified Excel file in default (that is Excel 2003) format
    workbook.save(self.dataDir + "Delete Column.xls")

print "Delete Column Successfully." 
```
